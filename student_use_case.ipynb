En utilisant, la dataframe que nous avons traités, Écrivez un programme qui demande à l'utilisateur d'entrer un code étudiant. (NB: vous pouvez utiliser la fonction input() et N'oubliez pas d'afficher un message d'error si le code de l'etudiant n'existe pas)

En utilisant le code étudiant saisi, affichez les informations suivantes de l'étudiant :

    Son nom et prenom
    Son email.
    Son sexe.
    Sa date de naissance.
    Sa spécialité.
    Le nombre de cours ratés.
    La mention obtenue.


En outre, pour la spécialité de l'étudiant saisi, affichez :

    Le nombre maximum de cours ratés parmi tous les étudiants de cette spécialité.
    
    Le moyenne de cours ratés parmi tous les étudiants de cette spécialité.

    Le nombre d'étudiants ayant obtenu une mention faible, passable, bien et très bien dans cette spécialité.



```python
def _check_student_code(
    data: pd.DataFrame, # students iformation
    id_etudiant: int # code etudiant
)-> bool:
    """check student Id"""
    
    if id_etudiant in data['id_etudiant'].values:
        return True

    return False
```

```python
def saisir_code_etudiant(
    data: pd.DataFrame, # students information
)-> None:

  # recuperer l'id par le user
  id_et = int(input("Entrer l'id de l'etudiant: "))

  if _check_student_code(data=data, id_etudiant=id_et):

    print(f"\nle code etudiant '{id_et}' existe. \n")

    # filtrer la data par le code etudiant
    student_data = data[data.id_etudiant == id_et]

    # afficher les info de l'etudiant
    print(f"**Nom et Prenom: {student_data.nom_etudiant.values[0]} {student_data.prenom_etudiant.values[0]}\n")
    print(f"**Email: {student_data.email.values[0]}\n")
    print(f"**Sex: {student_data.sex.values[0]}\n")
    print(f"**Date de Naissance: {student_data.date_naissance.values[0]}\n")
    print(f"**Mention: {student_data.mention.values[0]}\n")
    print(f"**Specialité: {student_data.specialite.values[0]}\n")
    print(f"**Cours_rate: {student_data.cours_rate.values[0]}\n")

    # filtrer la data par la specialité de l'etudiant
    specialité_etudiant = data[data.specialite == student_data.specialite.values[0]]

    # la moyenne de cours ratés
    moy_cours_rate = specialité_etudiant["cours_rate"].mean()
    print("**La moyenne de cours ratés: ",moy_cours_rate,"\n")

    # le nombre max de cours raté
    max_cours_rate = specialité_etudiant["cours_rate"].max()
    print("**Max de cours ratés: ",max_cours_rate,"\n")

    # le nombre d'etudiant par mention
    mentions = specialité_etudiant["mention"].value_counts().to_dict()
    print("**Le nombre d'étudiants par mentions:\n")
    print(mentions)

  else:
      print(f"Le code etudiant '{id_et}' n'existe pas.")
```

```python
saisir_code_etudiant(data)
```

## Data Imputation

Lorsque nous avons des valeurs manquantes, nous avons plusieurs choix à faire :
- Supprimer les lignes/colonnes
- Imputation des données

### Frais scolarite

```python
data[data["frais_scolarite"] != data["frais_scolarite"]] #  on peut utiliser cette methode data[data["frais_scolarite"].isnull()]
```

Pour l'imputation on peut utiliser plusieur technique comme :



*   La moyenne: La moyenne arithmétique est la somme des valeurs de la variable divisée par le nombre d’individus. (en python on utilise .mean())
*   Le mode : Le mode est la valeur la plus fréquente dans un échantillon. (en python en utilise .mode())
*   la médiane : La valeur centrale d'une série statistique. (en python en utilise .median())



```python
# fillna est une fonction pour appliquer l'imputation

data.fillna(
    {
        'frais_scolarite':data.frais_scolarite.mean(),
     }
    ,inplace=True)

# autre ecriture
#data = data.fillna({'frais_scolarite':data.frais_scolarite.mean()})
```

```python
# il ya plus de valeurs nan
data[data["frais_scolarite"] != data["frais_scolarite"]]
```
