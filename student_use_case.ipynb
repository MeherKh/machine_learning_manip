## Supprimer rows and columns

### supprimer les columns

#### methode 1

```python
data[data.columns[data.isnull().mean() < 0.5]] # supp les columns qui > 50 % des valeurs nulles
```

#### methode 2

```python
#Calculer le pourcentage de valeurs nulles dans chaque colonne
null_percentage = data.isnull().mean() * 100
null_percentage
```

```python
# Rechercher les colonnes avec plus de 50 % de valeurs nulles
columns_to_drop = null_percentage[null_percentage > 50].index.tolist()
columns_to_drop
```

```python
data = data.drop(columns=columns_to_drop) # ou bien data.drop(columns_to_drop, axis = 1) , axis = 1 c'est pour les colonnes
```
